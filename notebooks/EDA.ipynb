```python
import numpy as np
import pandas as pd
import os
# загрузка данных
df = pd.read_csv("/content/Аналитика персонала от IBM.csv")
df
```

```python
# первичный осмотр
# инфо (типы, пропуски)
df.info()

# описательная статистика (числа + категории)
display(df.describe().T)                 # числа
display(df.describe(include='object').T)  # категориальные
```

```python
# оценка полноты данных
# Completeness: пропуски по колонкам и в процентах
missing_count = df.isnull().sum()
missing_percent = (missing_count / len(df)) * 100
missing_df = pd.DataFrame({'missing_count': missing_count, 'missing_percent': missing_percent})
missing_df = missing_df.sort_values('missing_percent', ascending=False)
missing_df.head(20)
```

```python
# оценка уникальности записей
# Uniqueness — проверка ключей
# Предполагаемый идентификатор: EmployeeNumber
id_col = 'EmployeeNumber'
total = len(df)
unique_ids = df[id_col].nunique()
uniqueness = unique_ids / total

print(f"Total rows: {total}")
print(f"Unique {id_col}: {unique_ids}")
print(f"Uniqueness = {uniqueness:.4f}")

# проверить дубликаты строк целиком
dups = df.duplicated().sum()
print("Full row duplicates:", dups)
```

```python
# разделение признаков
# Numeric vs Categorical
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print("Numeric:", len(numeric_cols), numeric_cols)
print("Categorical:", len(categorical_cols), categorical_cols)
```

```python
# оценка выбросов — Outlier Ratio
# Outlier Ratio (IQR method) для каждого числового столбца
def outlier_ratio(series):
    if series.dropna().empty:
        return np.nan
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = series[(series < lower) | (series > upper)]
    return len(outliers) / len(series)

outlier_ratios = {col: outlier_ratio(df[col]) for col in numeric_cols}
outlier_df = pd.DataFrame.from_dict(outlier_ratios, orient='index', columns=['outlier_ratio'])
outlier_df = outlier_df.sort_values('outlier_ratio', ascending=False)
outlier_df.head(20)
```

```python
# Z-score method (опционально)
from scipy import stats
z_outlier_ratios = {}
for col in numeric_cols:
    s = df[col].dropna()
    if s.empty:
        z_outlier_ratios[col] = np.nan
        continue
    z = np.abs(stats.zscore(s))
    z_outlier_ratios[col] = (z > 3).sum() / len(s)

pd.DataFrame.from_dict(z_outlier_ratios, orient='index', columns=['z_outlier_ratio']).sort_values('z_outlier_ratio', ascending=False).head(10)
```

```python
# визуализация распределений и выбросов
# гистограммы для числовых признаков
import matplotlib.pyplot as plt
import seaborn as sns

def plot_numeric_grid(df, cols, ncols=3, figsize=(15, 12)):
    n = len(cols)
    nrows = (n + ncols - 1) // ncols
    plt.figure(figsize=figsize)
    for i, col in enumerate(cols):
        plt.subplot(nrows, ncols, i+1)
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(col)
    plt.tight_layout()

plot_numeric_grid(df, numeric_cols, ncols=3, figsize=(14, 18))
```

```python
# Boxplots (выбросы) — лаконично
plt.figure(figsize=(16, 6))
sns.boxplot(data=df[numeric_cols], orient='h')
plt.title("Boxplots for numeric features (horizontal)")
plt.show()
```

```python
# для некоторых ключевых признаков — сравнение по классу Attrition
if 'Attrition' in df.columns:
    plt.figure(figsize=(12,6))
    sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
    plt.title("MonthlyIncome by Attrition")
    plt.show()

    plt.figure(figsize=(12,6))
    sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
    plt.title("YearsAtCompany by Attrition")
    plt.show()
```

```python
# Категориальные распределения (countplots)
for col in ['Department', 'JobRole', 'BusinessTravel', 'OverTime']:
    if col in df.columns:
        plt.figure(figsize=(10,4))
        sns.countplot(y=col, data=df, order=df[col].value_counts().index)
        plt.title(col)
        plt.show()
```

```python
# корреляции
# 9. Корреляция (числовые)
corr = df[numeric_cols].corr(method='pearson')
plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap='coolwarm', center=0, annot=False)
plt.title("Pearson correlation matrix")
plt.show()
```

```python
# топ сильных пар (по абсолютному значению)
corr_unstacked = corr.abs().unstack()
# убираем самокорреляции
corr_unstacked = corr_unstacked[corr_unstacked < 1.0]
top_pairs = corr_unstacked.sort_values(ascending=False).drop_duplicates().head(20)
top_pairs
```

```python
# фичеинжиниринг
# бинарный target и соотношение опыта к возрасту
if 'Attrition' in df.columns:
    df['Attrition_Flag'] = df['Attrition'].map({'Yes':1, 'No':0})

# Experience ratio (делим с защитой от деления на ноль)
df['ExperienceRatio'] = df['TotalWorkingYears'] / df['Age'].replace(0, np.nan)
df[['TotalWorkingYears','Age','ExperienceRatio']].head()
```

```python
# КАСТОМНАЯ ВИЗУАЛИЗАЦИЯ В ЕДИНОМ СТИЛЕ

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Создаем директорию для графиков ПЕРЕД началом визуализации
os.makedirs('notebooks/visualizations', exist_ok=True)

# Настройка единого стиля для всех графиков
custom_style = {
    'style': 'whitegrid',
    'palette': 'viridis',
    'font_scale': 1.2,
    'rc': {
        'figure.figsize': (12, 8),
        'font.size': 12,
        'axes.titlesize': 14,
        'axes.labelsize': 12,
        'xtick.labelsize': 10,
        'ytick.labelsize': 10
    }
}

sns.set_theme(**custom_style)
plt.rcParams.update(custom_style['rc'])

# Кастомная палитра
custom_palette = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd",
                  "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"]
sns.set_palette(custom_palette)
```

```python
# 1. СЕТКА РАСПРЕДЕЛЕНИЙ КЛЮЧЕВЫХ ЧИСЛОВЫХ ПРИЗНАКОВ

key_numeric_features = ['Age', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany',
                       'DistanceFromHome', 'YearsWithCurrManager']

def plot_distribution_grid(df, cols, target=None, ncols=3, figsize=(16, 12)):
    """
    Сетка распределений с возможностью окраски по целевому признаку
    """
    n = len(cols)
    nrows = (n + ncols - 1) // ncols
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize)
    axes = axes.flatten() if n > 1 else [axes]

    for i, col in enumerate(cols):
        ax = axes[i]
        if target and target in df.columns:
            # Раздельное распределение по классам target
            for class_val in df[target].unique():
                subset = df[df[target] == class_val]
                sns.histplot(subset[col].dropna(), kde=True, label=class_val,
                           alpha=0.7, ax=ax)
            ax.legend(title=target)
        else:
            # Обычное распределение
            sns.histplot(df[col].dropna(), kde=True, ax=ax)

        ax.set_title(f'Распределение {col}', fontweight='bold')
        ax.set_xlabel(col)
        ax.set_ylabel('Частота')

    # Скрываем пустые subplots
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    return fig

# Распределения ключевых признаков
fig1 = plot_distribution_grid(df, key_numeric_features)
plt.suptitle('Распределения ключевых числовых признаков', y=1.02, fontsize=16, fontweight='bold')
plt.savefig('notebooks/visualizations/1_numeric_distributions.png', bbox_inches='tight', dpi=300)
plt.show()

# Распределения с разделением по Attrition
if 'Attrition' in df.columns:
    fig2 = plot_distribution_grid(df, key_numeric_features, target='Attrition')
    plt.suptitle('Распределения по признаку Attrition', y=1.02, fontsize=16, fontweight='bold')
    plt.savefig('notebooks/visualizations/2_attrition_distributions.png', bbox_inches='tight', dpi=300)
    plt.show()
```

```python
# 2. СЕТКА БОКСПЛОТОВ ДЛЯ ВЫБРОСОВ

def plot_boxplot_grid(df, cols, target=None, ncols=3, figsize=(16, 14)):
    """
    Сетка боксплотов для анализа выбросов
    """
    n = len(cols)
    nrows = (n + ncols - 1) // ncols
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize)
    axes = axes.flatten() if n > 1 else [axes]

    for i, col in enumerate(cols):
        ax = axes[i]
        if target and target in df.columns:
            sns.boxplot(data=df, x=target, y=col, ax=ax)
            ax.set_title(f'{col} по {target}', fontweight='bold')
        else:
            sns.boxplot(data=df, y=col, ax=ax)
            ax.set_title(f'Боксплот {col}', fontweight='bold')
        ax.tick_params(axis='x', rotation=45)

    # Скрываем пустые subplots
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    return fig

# Боксплоты ключевых признаков
fig3 = plot_boxplot_grid(df, key_numeric_features[:6])
plt.suptitle('Анализ выбросов в ключевых признаках', y=1.02, fontsize=16, fontweight='bold')
plt.savefig('notebooks/visualizations/3_boxplots_numeric.png', bbox_inches='tight', dpi=300)
plt.show()

# Боксплоты по Attrition
if 'Attrition' in df.columns:
    fig4 = plot_boxplot_grid(df, key_numeric_features[:6], target='Attrition')
    plt.suptitle('Распределения признаков по Attrition', y=1.02, fontsize=16, fontweight='bold')
    plt.savefig('notebooks/visualizations/4_boxplots_attrition.png', bbox_inches='tight', dpi=300)
    plt.show()
```

```python
# 3. СЕТКА КАТЕГОРИАЛЬНЫХ ПРИЗНАКОВ

categorical_features = ['Department', 'JobRole', 'BusinessTravel', 'OverTime',
                       'EducationField', 'Gender', 'MaritalStatus']

def plot_categorical_grid(df, cols, target=None, ncols=3, figsize=(18, 15)):
    """
    Сетка для категориальных признаков
    """
    n = len(cols)
    nrows = (n + ncols - 1) // ncols
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize)
    axes = axes.flatten() if n > 1 else [axes]

    for i, col in enumerate(cols):
        ax = axes[i]
        if target and target in df.columns:
            # Stacked bar plot с целевой переменной
            ct = pd.crosstab(df[col], df[target], normalize='index') * 100
            ct.plot(kind='bar', stacked=True, ax=ax, color=custom_palette[:len(df[target].unique())])
            ax.set_ylabel('Процент (%)')
            ax.legend(title=target, bbox_to_anchor=(1.05, 1), loc='upper left')
        else:
            # Обычный countplot
            sns.countplot(data=df, y=col, ax=ax,
                         order=df[col].value_counts().index)
            ax.set_ylabel('')

        ax.set_title(f'{col}', fontweight='bold')
        ax.tick_params(axis='x', rotation=0)
        ax.tick_params(axis='y', rotation=0)

    # Скрываем пустые subplots
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    return fig

# Категориальные распределения
fig5 = plot_categorical_grid(df, categorical_features[:6])
plt.suptitle('Распределение категориальных признаков', y=1.02, fontsize=16, fontweight='bold')
plt.savefig('notebooks/visualizations/5_categorical_distributions.png', bbox_inches='tight', dpi=300)
plt.show()

# Категориальные признаки по Attrition
if 'Attrition' in df.columns:
    fig6 = plot_categorical_grid(df, categorical_features[:6], target='Attrition')
    plt.suptitle('Категориальные признаки по Attrition', y=1.02, fontsize=16, fontweight='bold')
    plt.savefig('notebooks/visualizations/6_categorical_attrition.png', bbox_inches='tight', dpi=300)
    plt.show()
```

```python
# 4. МАТРИЦА РАССЕЯНИЯ ДЛЯ КЛЮЧЕВЫХ ПРИЗНАКОВ

# Убираем возможные пропуски для pairplot
pairplot_features = ['Age', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany', 'Attrition']
pairplot_df = df[pairplot_features].dropna()

if 'Attrition' in pairplot_df.columns:
    plt.figure(figsize=(14, 12))
    pair_plot = sns.pairplot(pairplot_df, hue='Attrition',
                            palette=custom_palette[:2],
                            diag_kind='kde',
                            plot_kws={'alpha': 0.7, 's': 50})
    pair_plot.fig.suptitle('Матрица рассеяния ключевых признаков по Attrition',
                          y=1.02, fontsize=16, fontweight='bold')
    pair_plot.savefig('notebooks/visualizations/7_pairplot_attrition.png', bbox_inches='tight', dpi=300)
    plt.show()
```

```python
# 5. ТЕПЛОВАЯ КАРТА КОРРЕЛЯЦИЙ С АННОТАЦИЕЙ

# Выбираем наиболее релевантные числовые признаки для корреляции
correlation_features = [col for col in key_numeric_features if col in df.columns]
correlation_features.extend(['Attrition_Flag'] if 'Attrition_Flag' in df.columns else [])

if len(correlation_features) > 1:
    corr_matrix = df[correlation_features].corr()

    # Маска для верхнего треугольника
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
                center=0, square=True, linewidths=0.5, cbar_kws={"shrink": .8})
    plt.title('Тепловая карта корреляций ключевых признаков', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.savefig('notebooks/visualizations/8_correlation_heatmap.png', bbox_inches='tight', dpi=300)
    plt.show()
```

```python
# 6. СЕТКА СВЯЗЕЙ КАТЕГОРИАЛЬНЫХ ПРИЗНАКОВ С ATTENTION


if 'Attrition' in df.columns:
    # Выбираем топ-4 категориальных признака для детального анализа
    top_categorical = ['Department', 'JobRole', 'OverTime', 'BusinessTravel']

    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    axes = axes.flatten()

    for i, col in enumerate(top_categorical):
        if i < len(axes):
            # Процент Attrition по категориям
            attrition_rate = df.groupby(col)['Attrition'].apply(
                lambda x: (x == 'Yes').mean() * 100
            ).sort_values(ascending=False)

            sns.barplot(x=attrition_rate.values, y=attrition_rate.index,
                       ax=axes[i], palette='viridis')
            axes[i].set_title(f'Процент Attrition по {col}', fontweight='bold')
            axes[i].set_xlabel('Процент Attrition (%)')
            axes[i].set_ylabel('')

    plt.suptitle('Анализ процента Attrition по категориальным признакам',
                fontsize=16, fontweight='bold', y=0.98)
    plt.tight_layout()
    plt.savefig('notebooks/visualizations/9_attrition_rates.png', bbox_inches='tight', dpi=300)
    plt.show()
```
