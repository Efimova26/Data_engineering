```python
import numpy as np
import pandas as pd
import os
# загрузка данных
df = pd.read_csv("/content/Аналитика персонала от IBM.csv")
df
```

```python
# первичный осмотр
# инфо (типы, пропуски)
df.info()

# описательная статистика (числа + категории)
display(df.describe().T)                 # числа
display(df.describe(include='object').T)  # категориальные
```

```python
# оценка полноты данных
# Completeness: пропуски по колонкам и в процентах
missing_count = df.isnull().sum()
missing_percent = (missing_count / len(df)) * 100
missing_df = pd.DataFrame({'missing_count': missing_count, 'missing_percent': missing_percent})
missing_df = missing_df.sort_values('missing_percent', ascending=False)
missing_df.head(20)
```

```python
# оценка уникальности записей
# Uniqueness — проверка ключей
# Предполагаемый идентификатор: EmployeeNumber
id_col = 'EmployeeNumber'
total = len(df)
unique_ids = df[id_col].nunique()
uniqueness = unique_ids / total

print(f"Total rows: {total}")
print(f"Unique {id_col}: {unique_ids}")
print(f"Uniqueness = {uniqueness:.4f}")

# проверить дубликаты строк целиком
dups = df.duplicated().sum()
print("Full row duplicates:", dups)
```

```python
# разделение признаков
# Numeric vs Categorical
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print("Numeric:", len(numeric_cols), numeric_cols)
print("Categorical:", len(categorical_cols), categorical_cols)
```

```python
# оценка выбросов — Outlier Ratio
# Outlier Ratio (IQR method) для каждого числового столбца
def outlier_ratio(series):
    if series.dropna().empty:
        return np.nan
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = series[(series < lower) | (series > upper)]
    return len(outliers) / len(series)

outlier_ratios = {col: outlier_ratio(df[col]) for col in numeric_cols}
outlier_df = pd.DataFrame.from_dict(outlier_ratios, orient='index', columns=['outlier_ratio'])
outlier_df = outlier_df.sort_values('outlier_ratio', ascending=False)
outlier_df.head(20)
```

```python
# Z-score method (опционально)
from scipy import stats
z_outlier_ratios = {}
for col in numeric_cols:
    s = df[col].dropna()
    if s.empty:
        z_outlier_ratios[col] = np.nan
        continue
    z = np.abs(stats.zscore(s))
    z_outlier_ratios[col] = (z > 3).sum() / len(s)

pd.DataFrame.from_dict(z_outlier_ratios, orient='index', columns=['z_outlier_ratio']).sort_values('z_outlier_ratio', ascending=False).head(10)
```

```python
# визуализация распределений и выбросов
# гистограммы для числовых признаков
import matplotlib.pyplot as plt
import seaborn as sns

def plot_numeric_grid(df, cols, ncols=3, figsize=(15, 12)):
    n = len(cols)
    nrows = (n + ncols - 1) // ncols
    plt.figure(figsize=figsize)
    for i, col in enumerate(cols):
        plt.subplot(nrows, ncols, i+1)
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(col)
    plt.tight_layout()

plot_numeric_grid(df, numeric_cols, ncols=3, figsize=(14, 18))
```

```python
# Boxplots (выбросы) — лаконично
plt.figure(figsize=(16, 6))
sns.boxplot(data=df[numeric_cols], orient='h')
plt.title("Boxplots for numeric features (horizontal)")
plt.show()
```

```python
# для некоторых ключевых признаков — сравнение по классу Attrition
if 'Attrition' in df.columns:
    plt.figure(figsize=(12,6))
    sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
    plt.title("MonthlyIncome by Attrition")
    plt.show()

    plt.figure(figsize=(12,6))
    sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
    plt.title("YearsAtCompany by Attrition")
    plt.show()
```

```python
# Категориальные распределения (countplots)
for col in ['Department', 'JobRole', 'BusinessTravel', 'OverTime']:
    if col in df.columns:
        plt.figure(figsize=(10,4))
        sns.countplot(y=col, data=df, order=df[col].value_counts().index)
        plt.title(col)
        plt.show()
```

```python
# корреляции
# 9. Корреляция (числовые)
corr = df[numeric_cols].corr(method='pearson')
plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap='coolwarm', center=0, annot=False)
plt.title("Pearson correlation matrix")
plt.show()
```

```python
# топ сильных пар (по абсолютному значению)
corr_unstacked = corr.abs().unstack()
# убираем самокорреляции
corr_unstacked = corr_unstacked[corr_unstacked < 1.0]
top_pairs = corr_unstacked.sort_values(ascending=False).drop_duplicates().head(20)
top_pairs
```

```python
# фичеинжиниринг
# бинарный target и соотношение опыта к возрасту
if 'Attrition' in df.columns:
    df['Attrition_Flag'] = df['Attrition'].map({'Yes':1, 'No':0})

# Experience ratio (делим с защитой от деления на ноль)
df['ExperienceRatio'] = df['TotalWorkingYears'] / df['Age'].replace(0, np.nan)
df[['TotalWorkingYears','Age','ExperienceRatio']].head()
```
